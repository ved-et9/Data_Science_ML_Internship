```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
data=pd.read_csv('/content/iris_dataset.csv')
data
```

```python
plt.scatter(data['sepal_length'],data['sepal_width'])
# Name your axes
plt.xlabel('Lenght of sepal')
plt.ylabel('Width of sepal')
plt.show()
```

```python
x=data.iloc[:,1:3]
x
```

```python
kmeans=KMeans(2)
kmeans.fit(x)
```

```python
wcss=[]
for i in range(1,151):
  kmeans=KMeans(i)
  kmeans.fit(x)
  wcs_it=kmeans.inertia_
  wcss.append(wcs_it)
wcss
```

```python
number_clusters=range(1,151)
plt.plot(number_clusters,wcss)
plt.xlabel('no. of clusters')
plt.ylabel('wcss')
```

```python
kmeans=KMeans(2)
kmeans.fit(x)
```

```python
id_clust=kmeans.fit_predict(x)
id_clust
```

```python
data_with_clusters=data.copy()
data_with_clusters['Clusters']=id_clust
data_with_clusters
```

```python
plt.scatter(data_with_clusters['sepal_length'],data_with_clusters['sepal_width'], c= data_with_clusters ['Clusters'], cmap = 'rainbow')
```

#Standardize the variable

```python
from sklearn import preprocessing
x_scaled=preprocessing.scale(data)
x_scaled
```

#Clustering the scaled data

```python
kmeans_new=KMeans(2)
kmeans_new.fit(x_scaled)
```

```python
clusters_scaled=data.copy()
clusters_scaled['cluster_pred']=kmeans_new.fit_predict(x_scaled)
plt.scatter(clusters_scaled['sepal_length'], clusters_scaled['sepal_width'], c= clusters_scaled ['cluster_pred'], cmap = 'rainbow')
```

```python
wcss
for i in range (1,10):
    kmeans= KMeans(i)
    kmeans.fit(x_scaled)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)
wcss
```

```python
kmeans_new=KMeans(3)
kmeans_new.fit(x_scaled)
clusters_scaled=data.copy()
clusters_scaled['cluster_pred']=kmeans_new.fit_predict(x_scaled)
plt.scatter(clusters_scaled['sepal_length'], clusters_scaled['sepal_width'], c= clusters_scaled ['cluster_pred'], cmap = 'rainbow')
```
