```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
test=pd.read_csv('/content/2.03.+Test+dataset.csv')
test
```

```python
test['Admitted']=test['Admitted'].map({'Yes':1,'No':0})
test['Gender']=test['Gender'].map({'Female':1,'Male':0})
test
```


1.  use model to make predictions based on the test data
2.  compare those with the actual outcome
3.calculate accuracy
4.create confusion matrix



```python
y=test['Admitted']
x1=test[['SAT','Gender']]
x=sm.add_constant(x1)
reg_line=sm.Logit(y,x)
reg=reg_line.fit()
```

```python
x
```

stats model does not provide confusion matrix

```python
def confusion_matrix(data,actual_data,model):
  pred_values=model.predict(data)
  bins=np.array([0,0.5,1])
  cm=np.histogram2d(actual_data,pred_values,bins=bins)[0]
  accuracy=(cm[0,0]+cm[1,1])/cm.sum()
  return cm,accuracy

```

```python
cm=confusion_matrix(x,y,reg)
cm
```
